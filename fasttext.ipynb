```python
import pandas as pd
import fasttext

# Load the CSV file
df = pd.read_csv('cleaned_PS_train.csv')
print(len(df))
# Preprocess the data for FastText
def preprocess_for_fasttext(content, label):
    return f"__label__{label} {content}"

# Apply preprocessing
df['fasttext_format'] = df.apply(lambda row: preprocess_for_fasttext(row['content'], row['labels']), axis=1)

# Save to a text file in FastText format
train_file = 'train.txt'
df['fasttext_format'].to_csv(train_file, index=False, header=False)
```

```python
import pandas as pd
import fasttext

# Load the CSV file
df = pd.read_csv('cleaned_PS_dev.csv')
print(len(df))
# Preprocess the data for FastText
def preprocess_for_fasttext(content, label):
    return f"__label__{label} {content}"

# Apply preprocessing
df['fasttext_format'] = df.apply(lambda row: preprocess_for_fasttext(row['content'], row['labels']), axis=1)

test_file = 'test.txt'
df['fasttext_format'].to_csv(test_file, index=False, header=False)
```

```python
# Train FastText model
model = fasttext.train_supervised(
    input=train_file,
    lr=0.5,
    epoch=25,
    wordNgrams=5,
    dim=500
)

model.save_model("fasttext_model.bin")
```

```python
print(model.words)
print(model.labels)
```

```python
result = model.test('test.txt')

# Print the results
print(result)
```

```python
ngrams = [2,3,5]
lr = [0.5,0.25,0.05]
epoch = [10,25,50]
dims = [200,300,500]

for i in lr:
    for j in ngrams:
        for k in epoch:
            for l in dims:
                model = fasttext.train_supervised(
                    input=train_file,
                    lr=i,
                    epoch=k,
                    wordNgrams=k,
                    dim=l
                )

                result = model.test('test.txt')
                print(f"For lr:{i} ngrams:{j} epochs:{k} dims:{l} results are {result}")
```
